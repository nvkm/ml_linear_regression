```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df  = pd.read_csv("tvmarketing.csv")
```

```python
df.head(10)
```

```python
def normalized(a, axis=-1, order=2):
    l2 = np.atleast_1d(np.linalg.norm(a, order, axis))
    l2[l2==0] = 1
    return a / np.expand_dims(l2, axis)
```

```python
x = normalized(df['TV'].iloc[:].values)[0]
y = normalized(df['Sales'].iloc[:].values)[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
w = np.random.randn()
b = np.random.randn()
step = 100000
eta = 0.05
n = len(X_train)
```

```python
for i in range(step):
    pred_y = w*X_train + b
    cost = (1/n) * sum((pred_y - y_train)**2)
    grad_w = (2/n) * np.matmul(pred_y - y_train, X_train.transpose())
    grad_b = (2/n) * sum(pred_y - y_train)
    w = w - eta * grad_w
    b = b - eta * grad_b
    print(cost)

    
```

```python
plt.scatter(X_test,y_test)
axes = plt.gca()
x_vals = np.array(axes.get_xlim())
y_vals = b + w * x_vals
plt.plot(x_vals, y_vals, color='red')
plt.show()
```
